This notebook reads the **cleaned dataset** and produces simple visuals:
1) Channel split over time  
2) Territory leaderboard  
3) Product/SKU contribution  
4) Promotion vs. Non‑Promo uplift  
5) Repeat purchase frequency  


## 1) Setup & Load

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.rcParams["figure.figsize"] = (9, 4.8)
plt.rcParams["axes.grid"] = True

# Primary root (Windows path you provided)
forced_root = Path(r"C:\Users\inchr\Downloads\Capstone Associate Data Analyst\omnichannel-growth-engine")
if forced_root.exists():
    BASE_DIR = forced_root
else:
    here = Path.cwd()
    BASE_DIR = here if (here / "clean").exists() and (here / "notebooks").exists() else here.parent

CLEAN_PATH = BASE_DIR / "clean" / "cleaned_fmcg_omnichannel_sales.csv"
assert CLEAN_PATH.exists(), f"Cleaned file not found at: {CLEAN_PATH}"

df = pd.read_csv(CLEAN_PATH, low_memory=False, parse_dates=[c for c in ["Date","Week_Start","Order_Date"] if c in pd.read_csv(CLEAN_PATH, nrows=0).columns])

print("✅ Loaded cleaned data:", df.shape)
print("From:", CLEAN_PATH)
df.head(3)
```

## 2) Normalize essential columns

```python

def first_present(candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None

col_date     = first_present(["Date","Week_Start","Order_Date","Week"])
col_channel  = first_present(["Channel","channel"])
col_terr     = first_present(["Territory","Region","Area"])
col_cat      = first_present(["Category","Product_Category","SKU_Category"])
col_revenue  = first_present(["Revenue","Sales","Net_Sales"])
col_units    = first_present(["Units","Quantity","Qty"])
col_order    = first_present(["Order_ID","OrderId","Txn_ID"])
col_customer = first_present(["Customer_ID","CustomerId","Cust_ID"])
col_promo    = first_present(["Promo_Flag","Promotion_Flag","Promo","Is_Promo"])

needed = [col_date, col_channel, col_terr, col_cat, col_revenue, col_units, col_order, col_customer]
missing_named = [n for n,v in zip(["Date","Channel","Territory","Category","Revenue","Units","Order_ID","Customer_ID"], needed) if v is None]
assert all(needed), f"Missing required columns: {missing_named}. Columns available: {list(df.columns)}"

# Types & cleaning
df[col_date] = pd.to_datetime(df[col_date], errors="coerce")
df[col_channel]  = df[col_channel].astype(str).str.strip().str.title()
df[col_terr]     = df[col_terr].astype(str).str.strip().str.title()
df[col_cat]      = df[col_cat].astype(str).str.strip()
df[col_revenue]  = pd.to_numeric(df[col_revenue], errors="coerce")
df[col_units]    = pd.to_numeric(df[col_units], errors="coerce")
if col_promo is None:
    df["__Promo"] = 0
    col_promo = "__Promo"
else:
    df[col_promo] = pd.to_numeric(df[col_promo], errors="coerce").fillna(0).astype(int)

df = df.dropna(subset=[col_date, col_channel, col_terr, col_cat, col_revenue])
print("Columns in use:", {"date": col_date, "channel": col_channel, "territory": col_terr, "category": col_cat, "revenue": col_revenue, "units": col_units, "order": col_order, "customer": col_customer, "promo": col_promo})
print("Shape after normalization:", df.shape)
```

## 3) Channel split over time

```python

ts = df.groupby([col_date, col_channel])[col_revenue].sum().reset_index()
pivot_ts = ts.pivot(index=col_date, columns=col_channel, values=col_revenue).fillna(0).sort_index()

plt.figure()
for ch in pivot_ts.columns:
    plt.plot(pivot_ts.index, pivot_ts[ch], marker="o", linewidth=1.5, label=ch)
plt.title("Channel Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend(title="Channel", ncol=2, fontsize=8)
plt.tight_layout()
plt.show()

pivot_ts.tail()
```

## 4) Territory leaderboard

```python

terr = df.groupby(col_terr)[col_revenue].sum().sort_values(ascending=False)

plt.figure()
plt.barh(terr.index[::-1], terr.values[::-1])
plt.title("Territory Sales Leaderboard (Total Revenue)")
plt.xlabel("Revenue")
plt.tight_layout()
plt.show()

terr.to_frame("Total_Revenue")
```

## 5) Product / SKU contribution

```python

sku = df.groupby(col_cat)[col_revenue].sum().sort_values(ascending=False)
topn = 10 if len(sku) > 10 else len(sku)
sku_top = sku.head(topn)

plt.figure()
plt.barh(sku_top.index[::-1], sku_top.values[::-1])
plt.title(f"Top {topn} Categories by Revenue")
plt.xlabel("Revenue")
plt.tight_layout()
plt.show()

sku_top.to_frame("Revenue")
```

## 6) Promotion vs. Non‑Promo uplift

```python

order_rev = df.groupby([col_order, col_promo])[col_revenue].sum().reset_index()
avg_rev = order_rev.groupby(col_promo)[col_revenue].mean()
labels = ["Non‑Promo" if int(i)==0 else "Promo" for i in avg_rev.index]

plt.figure()
plt.bar(labels, avg_rev.values)
plt.title("Average Revenue per Order: Promo vs Non‑Promo")
plt.ylabel("Avg Revenue per Order")
plt.tight_layout()
plt.show()

if len(avg_rev) == 2 and 0 in avg_rev.index and 1 in avg_rev.index and avg_rev[0] > 0:
    uplift = (avg_rev[1]/avg_rev[0]-1)*100
    print(f"📈 Promo uplift vs Non‑Promo: {uplift:.1f}%")
else:
    print("ℹ️ Not enough data to compute uplift cleanly.")
avg_rev.to_frame("Avg_Order_Revenue")
```

## 7) Repeat purchase frequency

```python

cust_orders = df.groupby(col_customer)[col_order].nunique()

plt.figure()
plt.hist(cust_orders.values, bins=10)
plt.title("Repeat Purchase Frequency (Orders per Customer)")
plt.xlabel("Orders per Customer")
plt.ylabel("Customer Count")
plt.tight_layout()
plt.show()

print(f"🔁 % repeat buyers: {(cust_orders>1).mean()*100:.1f}%")
print(f"📊 Avg orders per customer: {cust_orders.mean():.2f}")
cust_orders.describe().to_frame("orders_per_customer")
```

## 8) Summary snapshot

```python

summary = {}
terr_sum = df.groupby(col_terr)[col_revenue].sum()
if not terr_sum.empty:
    summary["Top territory"] = terr_sum.idxmax()

ch_sum = df.groupby(col_channel)[col_revenue].sum()
if not ch_sum.empty:
    summary["Top channel"] = ch_sum.idxmax()

try:
    order_rev = df.groupby([col_order, col_promo])[col_revenue].sum().reset_index()
    avg_rev = order_rev.groupby(col_promo)[col_revenue].mean()
    if 0 in avg_rev.index and 1 in avg_rev.index and avg_rev[0] > 0:
        summary["Promo uplift (%)"] = round(float((avg_rev[1]/avg_rev[0]-1)*100), 1)
except Exception:
    summary["Promo uplift (%)"] = "n/a"

cust_orders = df.groupby(col_customer)[col_order].nunique()
summary["Repeat buyer rate (%)"] = round(float((cust_orders>1).mean()*100), 1)

print("Insights summary:")
for k,v in summary.items():
    print(f"- {k}: {v}")
```
